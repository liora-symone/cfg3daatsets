# Wikipedia Data Scraping

```python
import pandas as pd
```

## Ariana Grande

### Honeymoon Tour

```python
# GitHub CSV file link
github_url1 = 'http://raw.githubusercontent.com/liora-symone/cfg3datasets/main/ariana/Honeymoon_clean_v3.csv'

# Read CSV file into a DataFrame
honeymoon_merge = pd.read_csv(github_url1)

# Display the DataFrame
honeymoon_merge
```

```python
# Drop rows in the DataFrame where all values are null
honeymoon_merge = honeymoon_merge.dropna(axis=0, how='all')

# Display the DataFrame
honeymoon_merge
```

### Dangerous Woman Tour

```python
# GitHub CSV file link
github_url2 = 'http://raw.githubusercontent.com/liora-symone/cfg3datasets/main/ariana/dangerous_data_cleaned_v2.csv'

# Read CSV file into a DataFrame
dangerous_merge = pd.read_csv(github_url2)

# Display the DataFrame
dangerous_merge
```

### Sweetener Tour

```python
# GitHub CSV file link
github_url3 = 'http://raw.githubusercontent.com/liora-symone/cfg3datasets/main/ariana/cleaned_sweetener_data_v2.csv'

# Read CSV file into a DataFrame
sweetener_merge = pd.read_csv(github_url3)

# Display the DataFrame
sweetener_merge
```

```python
# Concatenate DataFrames vertically
ariana_merge = pd.concat([honeymoon_merge, dangerous_merge, sweetener_merge], ignore_index=True)

# Display the concatenated DataFrame
ariana_merge
```

### Merge Ariana & Coldplay data

```python
# GitHub CSV file link
github_url4 = 'https://raw.githubusercontent.com/liora-symone/cfg3datasets/main/coldplay/coldplay_merged_data_v1.csv'

# Read CSV file into a DataFrame
coldplay_merge = pd.read_csv(github_url4)

# Display the DataFrame
coldplay_merge
```

```python
# GitHub CSV file link
github_url5 = 'http://raw.githubusercontent.com/liora-symone/cfg3datasets/main/ed_sheeran/edsheeran_merged_data_v1.csv'

# Read CSV file into a DataFrame
edsheeran_merge = pd.read_csv(github_url5)

# Display the DataFrame
edsheeran_merge
```

```python
# Concatenate DataFrames vertically
ariana_coldplay_ed = pd.concat([ariana_merge, coldplay_merge, edsheeran_merge], ignore_index=True)

# Display the concatenated DataFrame
ariana_coldplay_ed
```

```python
# GitHub CSV file link
github_url6 = 'http://raw.githubusercontent.com/liora-symone/cfg3datasets/main/justin/justin_merged_data_v2.csv'

# Read CSV file into a DataFrame
justin_merge = pd.read_csv(github_url6)

# Display the DataFrame
justin_merge
```

```python
# Concatenate DataFrames vertically
wikitourdata = pd.concat([ariana_coldplay_ed, justin_merge], ignore_index=True)

# Display the concatenated DataFrame
wikitourdata
```

```python
# FInd the unique values to check that all artists have been merged
unique_values = wikitourdata['artist_name'].unique()

# Display the unique values
print(unique_values)
```

```python
# Save the DataFrame to a CSV file
wikitourdata.to_csv('wikitourdata.csv', index=False)
```
