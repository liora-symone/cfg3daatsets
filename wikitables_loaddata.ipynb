## Ariana Grande

### Honeymoon Tour

```python
#!pip install scrapy
#!pip install html5lib

import json
import requests # library to handle requests
import pandas as pd
from bs4 import BeautifulSoup # library to parse HTML documents
```

```python
# Get the response in the form of html
wikiurl1='https://en.wikipedia.org/wiki/The_Honeymoon_Tour'
table_class="wikitable sortable jquery-tablesorter"
response=requests.get(wikiurl1)
print(response.status_code)
```

```python
# Parse data from the html into a beautifulsoup object
soup = BeautifulSoup(response.text, 'html.parser')
table=soup.find('table',{'class':"wikitable"})
```

```python
honeymoon_df=pd.read_html(str(table))
# convert list to dataframe
honeymoon_df=pd.DataFrame(honeymoon_df[0])
honeymoon_df
```

```python
# Save the DataFrame to a CSV file
honeymoon_df.to_csv('honeymoon_data.csv', index=False)
```

### Ghost Stories Tour (incomplete)

```python
# GitHub CSV file link
github_csv_url = 'https://raw.githubusercontent.com/liora-symone/cfg3datasets/main/coldplay/ghoststories_data.csv'

# Read CSV file into a DataFrame
ghoststories_df = pd.read_csv(github_csv_url)

# Display the DataFrame
ghoststories_df
```

```python
# Calculate the total number of tickets sold and revenue
total_tickets_sold = 20000  # Update this value with the actual total number of tickets sold
total_revenue = 3090000

# Remove the $ sign from the 'revenue' column
ghoststories_df['revenue'] = ghoststories_df['revenue'].replace('[\$,]', '', regex=True).astype(float)

# Remove the $ sign from the 'average_price' column
ghoststories_df['average_price'] = ghoststories_df['average_price'].replace('[\$,]', '', regex=True).astype(float)

# Calculate the mean for each column
average_price_mean = ghoststories_df['average_price'].mean()
tickets_available_mean = ghoststories_df['tickets_available'].mean()
capacity_mean = ghoststories_df['capacity'].mean()

# Fill missing values with the mean
ghoststories_df['average_price'] = ghoststories_df['average_price'].fillna(average_price_mean)
ghoststories_df['tickets_available'] = ghoststories_df['tickets_available'].fillna(tickets_available_mean)
ghoststories_df['capacity'] = ghoststories_df['capacity'].fillna(capacity_mean)

# Display the DataFrame
ghoststories_df
```

```python
# Save the DataFrame to a CSV file
ghoststories_df.to_csv('ghoststories_data.csv', index=False)
```
